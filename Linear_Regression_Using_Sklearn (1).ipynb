#### Importing the libraries

```python
import pandas
from pandas import DataFrame
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

#### Loading the data into dataframe using pandas

```python
data = pandas.read_csv('cost_revenue_clean.csv')
print(data)
```

```python
data.describe()
```

```python
import seaborn as sns
```

```python
sns.set()
```

#### Obtaining variable distributions

```python
sns.distplot(data["production_budget_usd"])
```

```python
sns.distplot(data["worldwide_gross_usd"])
```

#### Checking for correlation

```python
correlation = data.corr()
print(correlation)
```

**Note:** In machine learning lingo, X is called feature/ independent variable and y is called the target value.

```python
X = DataFrame(data, columns=['production_budget_usd'])
y = DataFrame(data, columns=['worldwide_gross_usd'])
```

Above, we are putting an entire column in a variable using dataframe.

*Variable = DataFrame (Data , column = ['Column Name'])*

```python
plt.figure(figsize=(10,6))
plt.scatter(X, y, alpha=0.3)
plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide Gross $')
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
plt.show()
```

#### Training the model

```python
regression = LinearRegression()
regression.fit(X, y)
```

#### Slope coefficient:

```python
regression.coef_    # theta_1
```

#### Intercept

```python
#Intercept
regression.intercept_
```

#### Obtaining the best fit line

```python
plt.figure(figsize=(10,6))
plt.scatter(X, y, alpha=0.3)

# Adding the regression line here:
plt.plot(X, regression.predict(X), color='violet', linewidth=3)

plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide Gross $')
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
plt.show()
```

#### Evaluating the model

```python
#Getting r square from Regression
regression.score(X, y)
```
